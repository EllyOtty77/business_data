```python
import requests
import re
import pandas as pd
from bs4 import BeautifulSoup

url = 'https://www.digitalstore.co.ke/'
r = requests.get('https://www.digitalstore.co.ke/collections/laptops')

soup = BeautifulSoup(r.text, 'html')
prods = soup.find_all('a', {'class': 'product-item__title text--strong link'})
products = [prod.text.strip() for prod in prods]

prices = soup.find_all('span', {'class': 'price price--highlight'})
price_texts = [pr.text.strip() for pr in prices]

# Clean the price texts to contain only digits and decimal points, and format them to 2 decimal places
clean_prices = []
for price in price_texts:
    # Use regex to find all digits and decimal points
    clean_price = re.findall(r'\d+', price.replace(',', ''))
    if clean_price:
        # Join the digits, convert to float, and format to 2 decimal places
        clean_price = float(''.join(clean_price)) / 100  # Convert to float and assume the last two digits are the decimal part
        clean_prices.append(f'{clean_price:.2f}')  # Format to 2 decimal places
```

```python
data = {'Product': products, 'Price': clean_prices}
df = pd.DataFrame(data)
```

```python
df.to_csv('product_prices.csv', index=False)
pd.read_csv('product_prices.csv')
```
